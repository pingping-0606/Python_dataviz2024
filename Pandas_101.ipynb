#Quiz 9
-วนลูปนับว่ามีนิทานที่มีคำว่าแม่ หรือพ่อ ในชื่อเรื่อง ทั้งหมดกี่เรื่อง เรื่องอะไรบ้าง
-มีคำดังกล่าวในนิทานชีวิต หรือนิทานประจำถิ่นมากกว่ากัน
-คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่อง ว่าแม่ หรือพ่อมากที่สุด

````
count_mom_dad = 0
list_mom_dad_story = []
for index, row in folk_df.iterrows():
    if 'แม่' in row['title'] or 'พ่อ' in row['title']:
        count_mom_dad += 1
        list_mom_dad_story.append(row['title'])

print(f'มีนิทานที่มีคำว่าแม่หรือพ่อในชื่อเรื่องทั้งหมด {count_mom_dad} เรื่อง')
print('รายชื่อนิทาน:')
for story in list_mom_dad_story:
    print(story)
````

````
count_mom_dad_life = 0
count_mom_dad_local = 0

for index, row in folk_df.iterrows():
    if ('แม่' in row['title'] or 'พ่อ' in row['title']):
        if row['type'] == 'นิทานชีวิต':
            count_mom_dad_life += 1
        elif row['type'] == 'นิทานประจำถิ่น':
            count_mom_dad_local += 1

if count_mom_dad_life > count_mom_dad_local:
    print('มีคำว่าพ่อหรือแม่ในนิทานชีวิตมากกว่า')
elif count_mom_dad_local > count_mom_dad_life:
    print('มีคำว่าพ่อหรือแม่ในนิทานประจำถิ่นมากกว่า')
else:
    print('มีคำว่าพ่อหรือแม่ในนิทานชีวิตและนิทานประจำถิ่นเท่ากัน')
````

````
creator_counts = {}
for index, row in folk_df.iterrows():
    if 'แม่' in row['title'] or 'พ่อ' in row['title']:
        creator = row['creator']
        if creator in creator_counts:
            creator_counts[creator] += 1
        else:
            creator_counts[creator] = 1

max_count = 0
most_frequent_creator = ""
for creator, count in creator_counts.items():
    if count > max_count:
        max_count = count
        most_frequent_creator = creator

print(f"คนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่าแม่หรือพ่อมากที่สุดคือ: {most_frequent_creator} ({max_count} เรื่อง)")
````

#Quiz 9
-print ชื่อนิทานทั้งหมดที่มีคำว่า ปู่ ออกมา

````
# prompt: print ชื่อนิทานทั้งหมดที่มีคำว่า ปู่ ออกมาโดยใช้ iloc

count_grandpa = 0
list_grandpa_story = []
for i in range(folk_df.shape[0]):
    if 'ปู่' in folk_df.iloc[i,0]:
        count_grandpa += 1
        list_grandpa_story.append(folk_df.iloc[i,0])

print(f'มีนิทานที่มีคำว่าปู่ในชื่อเรื่องทั้งหมด {count_grandpa} เรื่อง')
print('รายชื่อนิทาน:')
for story in list_grandpa_story:
    print(story)
````
