```
import pandas as pd
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

# Parallel Coordinates

```
pd.plotting.parallel_coordinates(df, 'Name');
```

## à¸›à¸£à¸±à¸š parameter: color

```
pd.plotting.parallel_coordinates(df, 'Name', color=('#FFD700', '#C0C0C0', '#CD7F32')) #html color
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=((1,0.8,0), (0,0.7,0), (0.6,0,1))) #rgb value (r,g,b) | r,g,b -> [0,1]
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=('r', 'g', 'b')) #à¸•à¸±à¸§à¸¢à¹ˆà¸­à¸‚à¸­à¸‡à¸ªà¸µ R G B W C M Y K à¹à¸”à¸‡ à¹€à¸‚à¸µà¸¢à¸§ à¸™à¹‰à¸³à¹€à¸‡à¸´à¸™ à¸‚à¸²à¸§ /à¸Ÿà¹‰à¸² à¸¡à¹ˆà¸§à¸‡ à¹€à¸«à¸¥à¸·à¸­à¸‡ à¸”à¸³ (à¸ªà¸µà¸›à¸£à¸´à¹Šà¸™à¹€à¸•à¸­à¸£à¹Œ)
```

# à¹€à¸£à¸´à¹ˆà¸¡ Matplotlib

### à¹à¸›à¸¥à¸‡ data à¸ˆà¸²à¸ pandas dataframe à¹€à¸›à¹‡à¸™ numpy array

```
np_data = df.iloc[:,:-1].to_numpy()
np_data
```

# Visualize array data à¸”à¹‰à¸§à¸¢ pixel

```
import matplotlib
```

```
matplotlib.rcParams['figure.figsize']=[15,15] #à¸à¸³à¸«à¸™à¸”à¸‚à¸™à¸²à¸”à¸‚à¸­à¸‡ figure

plt.imshow(np_data[:50,:],interpolation='nearest',cmap='plasma') #cmap -> colormap
```

```
plt.imshow(np_data[50:100,:],interpolation='nearest')
```
